```python
def compare_weeks(start_date, end_date):
    week_1_start = start_date - timedelta(days=start_date.weekday())
    week_1_end = week_1_start + timedelta(days=6)
    week_2_start = end_date - timedelta(days=end_date.weekday())
    week_2_end = week_2_start + timedelta(days=6)
    return [(week_1_start,week_1_end), (week_2_start,week_2_end)]
```

```python
compare_time_ranges = compare_weeks(start_time, end_time)
compare_time_ranges
```

```python
df_points = get_tcx_points_df(fname).drop(columns=['activity', 'timestamp'])
df_points
```

```python
df_diff_1 = df_points.dropna(how='any').astype('float')
df_diff= df_diff_1.diff().dropna(how='any')
df_diff = df_diff.query(' seconds==1 ')
# n_rows_to_group = 10
# df_diff = df_diff.groupby(df_diff.index // n_rows_to_group).agg({'dist':'sum','seconds':'sum','alt':'sum'}).reset_index(drop=True)
df_diff['speed'] = df_diff['dist']/ df_diff['seconds']*10**-3*3600
df_diff['ele_perc'] = df_diff['alt']/df_diff['dist']*100
df_diff['ele_perc'] = df_diff['ele_perc'].fillna(0)
df_diff['ele_perc'].replace(np.inf, 0, inplace=True)
df_diff
```

```python
fig = px.scatter(df_diff, y='ele_perc', x='speed', trendline='ols')
fig.show()
```
